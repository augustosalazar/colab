## Resistencias
Desarrollar un algoritmo que calcula la resistencia total de un circuito que tiene dos resistencias en paralelo y una tercera en serie. Recordar que para encontrar la resistencia equivalente de dos resistencias en paraleo hay que usar la siguiente formula:

RT = 1 / ((1/R1) + (1/R2))

Mientras que para resistencias en serie solo hay que sumarlas.

Pista: encontrar primero la resistencia equivalente en paralelo y después sumarla a resistencia que está en serie.


````
print("suponga que la resistencia 1 y 2 estan en paralelo y la 3 en serie")
r1=int(input("Inserte valor en ohms de resistencia 1:"))
r2=int(input("Inserte valor en ohms de resistencia 2: "))
r3= int(input("Inserte valor en ohms de resistencia 3: "))
rt= (1/((1/r1)+(1/r2)))+r3
print("Resistencia total",rt)
````

## Cajero2

¿Cuántos billetes de cada denominación (50.000, 20.000 y 10.000) hay que darle al cliente para un valor requerido? asumir que el valor es divisible entre 10.000

````
#ref : https://www.w3schools.com/python/python_operators.asp
print("Digite la cantidad")
cantidad = int(input(""))
cantidadB50 = cantidad//50000
v20 = (cantidad-(cantidadB50*50000))
cantidadB20 = v20//20000
v10 = (v20-(cantidadB20*20000))
cantidadB10 = v10//10000
print("La cantidad de billetes de $50.000 es",cantidadB50)
print("La cantidad de billetes de $20.000 es",cantidadB20)
print("La cantidad de billetes de $10.000 es",cantidadB10)
````
