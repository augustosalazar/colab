# Vectores

Un vector es una estructura de datos unidimensional que sirve para almacenar datos. A diferencia de una variable normal, la cual solo puede guardar un solo dato a la vez, un vector puede almacenar más de un dato.
Las condiciones de almacenamiento son:   
-	Todos los datos tienen que ser del mismo tipo del vector, el tipo de datos del vector se define en el momento de la declaración.   
-	El máximo número de valores a almacenar esta dictado por el tamaño del vector, definido en el momento de la declaración.    

Es importante tener en cuenta que el tamaño del vector TIENE QUE SER CONOCIDO, no puede ser una variable.

Un vector se define en pseudocódigo del siguiente modo:   

**tipoDeDatos nombreDelVector[tamañoVector]**





Vamos a pensar en la siguiente vector llamado V que almancena solo números enteros:

|   	|   	|   	|   	|   	|   	|   	|   	|   	|   	|
|---	|---	|---	|---	|---	|---	|---	|---	|---	|---	|
|pos 0	|pos 1	|pos 2	|pos 3	|pos 4	|pos 5	|pos 6	|pos 7	|pos 8	|pos 9	|pos 5	|
|2	|5	|6	|8	|	|
|   	|   	|   	|   	|   	|

La definición es:

Entero V[10]

V tiene capacidad para 10 posiciones, aunque en el momento solo se están usando 4.

Para trabajar con los valores podemos:

Escribir v[1]  -> esto muestra un 5   
Escribir v[3]  -> esto muestra un 8   
Escribir v[11]  -> esto muestra un error, no hay posición 11   
Escribir v[4]  -> esto muestra un error, no hay nada en la posición 5   
Escribir v     -> esto muestra error, un vector no se puede escribir, solo las posiciones por separado.   
Leer v -> esto muestra error, un vector no se puede leer, solo las posiciones por separado.   
Leer v[8]  -> esto almacena un valor en la posición 8  

|   	|   	|   	|   	|   	|   	|   	|   	|   	|   	|
|---	|---	|---	|---	|---	|---	|---	|---	|---	|---	|
|pos 0	|pos 1	|pos 2	|pos 3	|pos 4	|pos 5	|pos 6	|pos 7	|pos 8	|pos 9	|pos 5	|
|2	|5	|6	|8	| | | | | 14 |	|
|   	|   	|   	|   	|   	|

Otras operaciones:

v[4] = v[0] + v[1] -> esto almacena el valor 7 en la posición 5

v[6] = v -> esto muestra un error, v no es una variable, es un vector

Una empresa maneja un sistema de información con tres vectores, en uno de ellos tiene los códigos de los productos (VC), en otro su precio unitario (VP) y en el último su inventario (VI). El vínculo entre estos tres vectores es el índice, eso quiere decir:
Si tenemos un producto cuyo código en VC[i] entonces podremos su precio unitario usando el mismo índice VP[i] y la cantidad disponible en VI[i].   

Diseñar un algoritmo que maneje el proceso de ventas (m clientes y n tipos de productos) con las siguientes características:
-	Un cliente solo puede pedir un tipo de producto
-	No se puede vender más productos que los disponibles en el inventario 
-	Si un cliente pide más productos que los disponibles se le dice adiós.
-	Después de cada cliente debemos mostrarle el valor a pagar
-	Al final del día se debe mostrar el código del producto que más unidades vendió. 


```
import random
# asume que se lee en orden
def leerRandVM(vector,numElementos,mult): 
  for i in range(numElementos):
    vector[i] = random.randrange(10*mult)
  return vector

def printVNombre(nombre, vector, size):
  print(nombre, end="\t")
  for j in range(size):
    if (vector[j] != None):
      print(vector[j],sep="",end="\t")
  print("")

def tieneRepetido(vector,tamano):
  for i in range (tamano):
    for j in range (tamano):
      if (vector[j] == vector[i] and i != j):
        return True
  return False

def existeEnVector(vector,numElementos,elemento):
  repetido = False
  for i in range(numElementos):
    if(vector[i] == elemento):
      repetido = True
  return repetido

def posicionEnVector(vector, size, elemento):
  for i in range (size):
    if (vector[i] == elemento):
      return i

def inicializarVector(vector, numElementos):
  for i in range (numElementos):
    vector[i] = 0
  return vector

def posMaxVector(vector,numElementos):
  max = 0 
  for i in range(numElementos):
    if (i == 0):
      max = vector[i]
      maxVect = i
    elif (vector[i]>max):
      max = vector[i]
      maxVect = i
  return maxVect


def posMinVector(vector,numElementos):
  min = 0 
  for i in range(numElementos):
    if (i == 0):
      min = vector[i]
      minVect = i
    elif (vector[i]<min):
      min = vector[i]
      minVect = i
  return minVect

##### MAIN #######

vc = [None]*100
vi = [None]*100
vp = [None]*100
vv = [None]*100
n = 10
m = 20
#n = int(input("digite n (cantidad de elementos)"))
#m = int(input("digite m (cantidad de clientes)"))
vv = inicializarVector(vv,n)
vc = leerRandVM(vc,n,10)
while (tieneRepetido(vc,n) == True):
  vc = leerRandVM(vc,n,10)
vi = leerRandVM(vi,n,1)
vp = leerRandVM(vp,n,100)

printVNombre("VC",vc,100)
printVNombre("VI",vi,100)
printVNombre("VP",vp,100)

for i in range(m):
  codProd = random.randrange(10*10)
  while (existeEnVector(vc,n,codProd) == False):
    codProd = random.randrange(10*10)
  cant = random.randrange(10)
  #print("El cliente",i,"pidio del producto con código",codProd,"cant",cant)
  pos = posicionEnVector(vc,n,codProd)
  if (vi[pos] > cant):
    vi[pos] = vi[pos] - cant
    vv[pos] = vv[pos] + cant
    #print("se puede vender con total",vp[pos]*cant)
  #else:
    #print("adios")

print('POST VENTA')
printVNombre("VC",vc,100)
printVNombre("VI",vi,100)
printVNombre("VP",vp,100)
printVNombre("VV",vv,100)
posMax = posMaxVector(vv,n)
print("El código del prod más vendido es",vc[posMax])
posMin = posMinVector(vv,n)
print("El código del prod menos vendido es",vc[posMin])
```
