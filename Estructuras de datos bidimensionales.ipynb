Implementar un sistema de información que tiene la siguiente estructura para n estudiantes:
-	En la primera columna (0) se tienen los códigos de los estudiantes
-	La segunda columna (1) tiene el primer parcial de cada estudiante
-	La tercera columna (2) tiene el segundo parcial de cada estudiante

El programa debe:
1.	Encontrar el promedio del primer parcial y colocarlo en la posición n,1
2.	Encontrar el promedio del segundo parcial y colocarlo en la posición n,2
3.	Encontrar la definitiva de cada estudiante y almacenarlo en la tercera columna.
4.	Dar el código del estudiante con el mayor promedio


Después de una captura de una noche estrellada se recogieron los datos en una matriz de NxN. Los científicos encontraron que el único modo de reconocer una estrella era buscando un 1 que estuviera rodeado de un 0 arriba, abajo, a la izquierda y a la derecha.   
Implementar un algoritmo para leer la matriz y decir cuántas estrellas tiene.

```
import random

def leerParcial(matriz,p,n):
  for i in range(n):
      matriz[i][p] = round(random.uniform(1,5), 1)
  return matriz

def promedioParcial(matriz,parcial,n):
  pr = 0
  for i in range (0,n):
    pr = pr + matriz[i][parcial]
  matriz[n][parcial] = round(pr/n,1)
  return matriz

def printM(m,f,c):
  for i in range(f):
    for j in range(c):
      print(m[i][j], end = '\t')
    print("")
  print("--------------------")


def definitiva(matriz,n,numeroParciales):
  for i in range(n):
    matriz[i][3] = 0
    for j in range(1,numeroParciales+1):
      matriz[i][3] = matriz[i][3] + matriz[i][j]
    matriz[i][3] =  round(matriz[i][3] / numeroParciales,1)
  return matriz

def mejorNota(m,n):
  mayor = 0
  ite = 0
  for i in range(n):
    if m[i][3] > mayor:
      mayor = m[i][3]
      ite = m[i][0]
  return ite

n = 20  # número de estudiantes
matriz = [[None for x in range(100)] for y in range(100)]

# llenando los códigos de los estudiantes
for i in range(n):
  matriz[i][0] = random.randint(0,100) 

matriz = leerParcial(matriz, 1, n) # la lectura tiene que ser para todos los n estudiantes de un parcial dado
matriz = leerParcial(matriz, 2, n)

printM(matriz,n,3)

matriz = promedioParcial(matriz, 1, n)
matriz = promedioParcial(matriz, 2, n)

printM(matriz,n+1,3)

numeroParciales = 2
matriz = definitiva(matriz,n, numeroParciales)

printM(matriz,n+1,4)

codMejorNota = mejorNota(matriz,n)

print(codMejorNota)
```

Una agencia automotriz distribuye quince modelos diferentes de coches y tiene diez vendedores. Se desea un programa que escriba un informe mensual de las ventas por vendedor y modelo, así como el número de automóviles vendidos por cada vendedor y el número total de cada modelo vendido por todos los vendedores. Asimismo, para entregar el premio al mejor vendedor, necesita saber cuál es el vendedor que más coches ha vendido.

filas son vendedores.  
columnas con modelos

```
import random

def printM(m,f,c):
  for i in range(f):
    for j in range(c+1):
      print(m[i][j], end = '\t')
    print("")
  print("--------------------")

def leerVentas(matriz,vendedores,modelos):
  for i in range(vendedores):
    for j in range(modelos):
      matriz[i][j+1] = random.randint(0,25)
  return matriz

def totalizarVentasVendedor(matriz, vendedores, modelos):
  for i in range(vendedores):
    suma=0
    for j in range(modelos):
      suma= suma + matriz[i][j+1]
    matriz[i][modelos+1] = suma
  return matriz

def totalizarVentasModelo(matriz, vendedores, modelos):
  for i in range(1,modelos+1):
    suma=0
    for j in range(vendedores):
      suma = suma + matriz[j][i]
    matriz[vendedores][i]= suma
  return matriz

def codigoVendedorMejorVentas(matriz, vendedores,modelos):
  mayor = 0
  mayori = 0
  for i in range (vendedores):
      if (matriz[i][modelos+1] > mayor):
          mayor = matriz[i][modelos+1]
          mayori = i
  print(matriz[mayori][0])


vendedores = 3  
modelos = 5

matriz = [[None for x in range(100)] for y in range(100)]

# llenando los códigos de los vendedores
for i in range(vendedores):
  matriz[i][0] = random.randint(0,100) 

printM(matriz, vendedores, modelos)

matriz = leerVentas(matriz,vendedores, modelos) ## llenar todas las ventas

printM(matriz, vendedores, modelos)

matriz = totalizarVentasVendedor(matriz, vendedores, modelos) # los totales colocarlos en la columna modelos + 1

printM(matriz, vendedores, modelos+1)

matriz = totalizarVentasModelo(matriz, vendedores, modelos) # los totales colocarlos en la fila vendedores + 1

printM(matriz, vendedores+1, modelos+1)

codV = codigoVendedorMejorVentas(matriz, vendedores, modelos)

print(codV)
```
