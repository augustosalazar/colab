# Condiciones simples
Un ciclo es una estructura algorítmica que nos permite ejecutar una acción hasta que una condición se cumpla. Esta condición puede ser tan sencilla como verificar que el número de iteraciones no sobrepase un limite puesto por el problema o una más compleja que dependa de múltiples variables y entradas de usuario.

Vamos a comenzar por la forma más simple, en la que vamos a realizar la acción un número N de veces. En este algoritmo usamos una variable como índice (i), esta va a comenzar en 1 y se va a incrementar después de cada iteración.

```
n = int(input("Digite el número de iteraciones "))

i = 1
print("valor inicial de n",n, "valor inicial de i",i)
print("Vamos a verificar si n",n, "es mayor o igual que i",i)
while (n>=i):
  print("(DENTRO) Voy a realizar acción ya que",n," es mayor o igual que",i)
  print("(DENTRO) A")
  print("(DENTRO) Ahora voy a incrementar i")
  i = i + 1
  print("(DENTRO) El nuevo valor de i",i)
  print("(DENTRO) -----------------------------------------------------------------------")
  print("Ahora se volvera a verificar si n",n," todavía es mayor o igual a i",i)
print("DONE")
  
```

## El día de mañana

Para N fechas (d m a) decir la fecha siguiente

```
n = int(input("Cuántas fechas vamos a trabajar? "))
i = 1

while(n>=i):
  dia = int(input('Día? '))
  mes = int(input('Mes? '))
  anio = int(input('Año '))

  if (dia < 30):
      dia = dia + 1
  else:
      dia = 1
      if (mes < 12):
          mes = mes + 1
      else:
          mes = 1
          anio = anio + 1
  print(dia,mes,anio,sep="/")
  i = i + 1
```

## Contador

En este tipo de ejercicios queremos contar el número de veces que una condición especifica se cumpla dentro del ciclo, en el siguiente ejemplo vamos a pedirle al usuario N números y contar cuántos de ellos son mayores que 10.

```
n = int(input("Digite la cantidad de números a evaluar "))
i = 1
contador = 0
while(n >= i):
  print("Digite el",i,"número")
  x = int(input())
  if (x > 10):
    contador = contador + 1
  i = i + 1
print("Tenemos",contador,"números mayores que 10")
```

En el siguiente ejemplo vamos a pedirle al usuario N números y contar cuántos de ellos son mayores que 10 y cuántos son mayores que 100.

```
n = int(input("Digite la cantidad de números a evaluar "))
i = 1
contador1 = 0
contador2 = 0
while(n >= i):
  print("Digite el",i,"número")
  x = int(input())
  if (x > 10):
    contador1 = contador1 + 1
  if (x > 100):
    contador2 = contador2 +1
  i = i + 1
print("Tenemos",contador1,"números mayores que 10")
print("Tenemos",contador2,"números mayores que 100")
```

## Promedio

Aquí queremos obtener el promedio de los valores ingresados por el usuario en cada iteración, primero vamos a mostrar un ejemplo simple:



```
n = int(input('Cuántas iteraciones quiere? '))
i = 1
promedio = 0
while(n >= i):
  print("Digite un número")
  x = float(input())
  promedio = promedio + x
  i = i + 1
promedio = promedio / n
print("El promedio es",promedio)
```

Para N números, encontrar el promedio de aquéllos mayor o iguales que 10

```
n = int(input('Cuántas iteraciones quiere? '))
i=1
c = 0
promedio = 0
while(n >= i):
  print("Digite un número")
  x = float(input())
  if (x >= 10):
    promedio = promedio + x
    c = c + 1
  i = i + 1

if (c == 0):
  promedio = 0
else:
  promedio = promedio / c

print("El promedio es",promedio)
```

## Mayor

En este ejercicio vamos a devolver el mayor entre los N números ingresados por el usuario, la estrategia es usar una variable llamada mayor que será inicializada en 0 (asumiendo que los valores son positivos) y después comparada con los valores ingresados.


```
n = int(input("Digite el número de iteraciones "))
i = 1
mayor = 0
while(n >= i):
  x = int(input("Digite un valor "))
  if (x > mayor):
    mayor = x
  i = i + 1
  
print("El mayor es",mayor)
```

Ahora vamos a implementar sin asumir nada de los valores

```
n = int(input("Digite el número de iteraciones "))
i = 1
mayorx = 0
while(n >= i):
  x = int(input("Digite un valor "))
  if (i == 1):
    mayorx = x
  else:
    if (x > mayorx):
      mayorx = x
  i = i + 1
  
print("El mayor es",mayorx)
```

En los casos en que no sabemos el rango de los valores que van a ser ingresados se inicializa el mayor con el primer dato ingresado.



## Menor

Este ejercicio sigue la misma formula para encontrar el menor de los valores, vamos a inicializar la variable menor en 100 asumiendo que el mayor valor que nos van a dar es 100.


```
n = int(input("Digite el número de iteraciones "))
i = 1
menor = 100
while(n >= i):
  x = int(input("Digite un valor "))
  if (x < menor):
    menor = x
  i = i + 1
  
print("El menor es",menor)
```

En los casos en que no sabemos el rango de los valores que van a ser ingresados se inicializa el menor con el primer dato ingresado.

# Condiciones complejas

Hasta ahora hemos trabajado con una condición sencilla en la que nos limitamos a verificar que el índice sea menor o igual al número de iteraciones esperadas. En los siguientes ejercicios vamos a ver algunos de los casos en que hay que trabajar con condiciones complejas.

## Buscar un subconjunto dentro de N elementos

Un ejemplo bastante común es en el cual tenemos un conjunto de N elementos sobre el que queremos encontrar un subconjunto de X elementos con alguna característica en común. Las salidas sería la confirmación si ese subconjunto fue encontrado.

Por ejemplo, dentro de un conjunto de N números buscar X números pares.


```
#primero tenemos que leer N (es la máxima cantidad de elemenos que podemos verificar)
n = int(input('Digite n '))
#x es la cantidad de elemenos pares que queremos encontrar
x = int(input('Digite x '))

#el algoritmo puede parar por dos cosas:
# 1. Encontramos los x elemenos
# 2. Terminamos de analizar lis n elemenos y no encontramos los x elemenos

#vamos primero a hacer la estructura básica

i = 1
contador = 0
while(n >= i and x > contador):
  valor = int(input('Digite un valor '))
  if (valor % 2 == 0):
    contador = contador + 1
    print(valor,'es par, llevamos',contador,'pares')
  i = i + 1
if (x == contador): 
  print('Encontramos los pares')
else:
  print('No encontramos los pares')


```

## Buscar N elementos dentro de un conjunto ilimitado  

Una variación del caso anterior, en el cuál nuestro ciclo ya no depende del índice sino de un contador que se incrementa ante la ocurrencia de una condición dentro de cada iteración.

Por ejemplo, solicitar números hasta que se encuentren N pares.


```
x = int(input("Digite cuántos pares quiere "))
c = 0
while( c < x ):
  w = int(input("Digite un valor "))
  if w % 2 == 0:
    c = c + 1
    print("Par encontrado, van",c)

print("Llegamos a",c)
```

## Iterar hasta que el usuario lo decida

En este caso vamos a realizar la acción hasta que el usuario decida terminar el ciclo, en este caso la condición debe depender de una entrada de un dato de entrada. Para la estructura del MientrasQue debemos inicializar ese dato de entrada antes de la primera iteración y realizar la lectura después de la misma.

Por ejemplo, solicitar números contando ocurrencias de pares e impares hasta que el usuario lo decida.


```
pares = 0
impares = 0
rta = 'si'
i = 1
while rta == 'si':
  x = int(input('Digite un número '))
  if (x % 2 == 0):
    pares = pares + 1
  else:
    impares = impares + 1
  if (i > 2):
    rta = input('Quiere continuar? (si/no) ')
    while(rta != 'si' and rta != 'no'):
      rta = input('Quiere continuar? (si/no) ')
  i = i + 1


print('pares',pares,'imapares',impares)

```

# Ciclo Para

Vamos a repasar la estructura del MientrasQue:

(1)	i = 1   
(2)	MientrasQue(n >= i)   
(3)	&nbsp;&nbsp;Escribir “acción “   
(4)	&nbsp;&nbsp;i = i + 1   
(5)	FinMientrasQue

Un ciclo Para es una versión reducida de un MientrasQue, es una solo línea vamos a hacer lo que hacíamos en las líneas (1), (2) y (3). Vamos a ver ese mismo ejemplo con un ciclo Para:

(1)	Para i = 1 hasta n saltando 1   
(2)	&nbsp;&nbsp;Escribir “acción “   
(3)	FinPara

Lo podemos leer de la siguiente manera “Para i que comienza en 1 hasta n de 1 en 1”. Esta sola línea inicializa el índice, le da su limite y describe de cuanto es el salto.

Es clave recordar que un MientrasQue podrá siempre reemplazar un Para, pero un Para no siempre puede reemplazar un MientrasQue. El Para solo funciona cuando sabemos cuantas iteraciones vamos a hacer, no funciona con condicionales complejos o entradas del usuario.


```
n = int(input("Digite n "))

for i in range(n):
  print(i)

print(".....")

for i in range(1,n):
  print(i)

print(".....")

for i in range(1,n+1):
  print(i)

print(".....")

for i in range(1,n+1,2):
  print(i)

print(".....")

for i in range(n,0):
  print(i)
```
