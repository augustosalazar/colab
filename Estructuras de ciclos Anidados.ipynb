# Ciclos anidados

El concepto de ciclos anidados es tan simple como complejo, en pocas palabras y como su nombre lo sugiere, es un ciclo con uno o más ciclos dentro de el. Vamos a usar un ejemplo para clarificar la idea:

Un día tiene 24 horas, el reloj va desde la hora 0 hasta la hora 23, veamos la implementación de un algoritmo que muestre cada hora durante un día:


```
for h in range(24):
  for m in range(60):
    print(h,m,sep=":")
```

```
from time import sleep
from IPython.display import clear_output

for h in range(24):
  for m in range(60):
    for s in range(60):
      print(h,m,s,sep=":")
      sleep(1)
      clear_output(wait=True)
```

```
from time import sleep
from IPython.display import clear_output

for h in range(24):
  for m in range(60):
    print(h,m,sep=":")
    sleep(60)
    clear_output(wait=True)
```

Ahora vamos a unirlos, claramente las horas se mueven más lento, por cada iteración del ciclo de las horas se ejecutan todas las iteraciones del ciclo de minutos, por lo cual el ciclo de las horas es el exterior y el de los minutos el interior. 
Lo único que hay que tener en cuenta es que ambos ciclos no pueden tener el mismo índice.

```
from time import sleep
from IPython.display import clear_output
x = 5
for m in range (x, 0, -1):
  for s in range (59,0,-1):
    print(m,s)
    sleep(1)
    clear_output(wait=True)
```

## Sumatoria

Este tipo de ejercicio comienza con la variable que terminara con el resultado inicializada en cero, en cada una de las iteraciones se acumula en esa variable un nuevo valor.


 $ S = \displaystyle\sum_{i=1}^{n} i$





```
n = int(input())
s = 0
for i in range(1,n+1):
  s = s + i
print('El valor final de s',s)
```

```
def miSum(n):
  s = 0
  for i in range(1,n+1):
    s = s + i
  return s

n = int(input())
print('El valor final de s',miSum(n))
```

```
j = int(input())
n = int(input())

s = 0
for i in range(j,n+1):
  s = s + i
print('El valor final de s',s)
```

## Productos

Una productoria tiene la misma estructura de una sumatoria, la única diferencia es que el resultado se inicializa en 1 y en cada iteración se multiplica y no se suma.


 $ S = \displaystyle\prod_{i=1}^{n} i$

```
n = int(input())
s = 1
for i in range(1,n+1):
  s = s * i
print('El valor final de s',s)
```

## Exponencial

```
n = int(input())
x = int(input())
e = 1
for i in range(x):
  e = e * n
print('El valor final de s',e)
print('El valor final de s',n**x)
```

### Sumatorias y Productos

Ahora vamos a ver un ejercicio un poco más complejo:

 $ S = \displaystyle\prod_{i=1}^{n} (\displaystyle\sum_{j=1}^{i} (i*\displaystyle\sum_{k=1}^{j} i+j) + \displaystyle\sum_{z=1}^{m} z)$

Leer n,m    
s = 1  // aquí vamos a guardar el resultado final    
Para i = 1 hasta n con saltos de 1. // este ciclo es del primer producto    
&nbsp;&nbsp;&nbsp;&nbsp;p = 0 // aquí vamos a guardar el resultado de la primera sumatoria    
&nbsp;&nbsp;&nbsp;&nbsp;Para j= 1 hasta i con saltos de 1. // este ciclo de la primera sumatoria    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;r = 0 // aquí vamos a guardar el resultado de la segunda sumatoria    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Para k= 1 hasta j con saltos de 1. // este ciclo de la segunda sumatoria    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;r = r + (i + j)    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;finPara // fin de ciclo de la segunda sumatoria    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;p = p + (i * r)    
&nbsp;&nbsp;&nbsp;&nbsp;finPara // fin de ciclo de la primera sumatoria    
&nbsp;&nbsp;&nbsp;&nbsp;w = 0 // aquí vamos a guardar el resultado de la tercera sumatoria    
&nbsp;&nbsp;&nbsp;&nbsp;Para z= 1 hasta m con saltos de 1. // este ciclo de la tercera sumatoria    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w = w + z    
&nbsp;&nbsp;&nbsp;&nbsp;finPara // fin de ciclo de la tercera sumatoria    
&nbsp;&nbsp;&nbsp;&nbsp;s = s * (p + w)    
finPara // fin de ciclo del primer producto    


```
n = int(input())
m = int(input())

p = 1
for i in range(1,n+1):
  s=0
  for j in range(1,i+1):
    q = 0
    for k in range(1,j+1):
      q = q + i + j
    s = s + (i*q)
  w = 0
  for z in range(1,m+1):
    w = w + z
  p = p * (s + w)

print('La respuesta es:',p)
```
