# Estructuras condicionales

Una estructura condicional cambia el flujo del algoritmo.

Los condicionales son una o más expresiones lógicas

Es una expresión alrededor de un operador de igualdad
==   igual
!=   distinto
<    menor
<=   menor o igual
>    mayor
>=   mayor o igual

Varias expresiones lógicas se pueden unir con operadores lógicos.  
and (conjunción)  
or (disyunción)


## Estructura básica (SI-FinSi)

El siguiente código en Java muestra el uso un condicional que solo tiene instrucciones en caso de que la condición sea verdadera


```
n = int(input())
print("Estoy antes del condicional")
if (n > 10):
  print("Estoy dentro del condicional")
print("Estoy fuera del condicional")
```

```
n = int(input())
print("Estoy antes del condicional") # siempre

if (n > 10):
  print("Estoy en el caso positivo del condicional") # 11

else:
  print("Estoy en el caso negativo del condicional") # 4, 5 (no se ve la linea 11)

  if (n > 5):
    print("Estoy en el positivo del negativo") # 6


print("Estoy fuera del condicional") # siempre
```

```
n = int(input())
print("Estoy antes del condicional")
if (n > 10):
  print("Estoy en el caso positivo del condicional")
else:
  print("Estoy en el caso negativo del condicional")
  if (n > 5):
    print("Estoy en el positivo del negativo")
print("Estoy fuera del condicional")
```

```
%%writefile Condicional1.java
import java.util.Scanner; 
class Condicional1 
{ 
    public static void main(String args[]) 
    { 
      // nuestra herramienta para leer datos, siempre la vamos a usar
    Scanner s = new Scanner(System.in); 
     
      // definimos las variables
    int n;
     
      // adquirimos los datos
    System.out.println("Digite un número entero"); 
    n = s.nextInt();
      
      if (n > 0){  // nuestro condicional
        System.out.println("El número es pósitivo"); // esta instrucción solo se ejecuta si la expresión lógica es verdadera
      } 
     
      System.out.println("Gracias por usar el programa!"); // esta instrucción siempre se ejecuta
    } 
} 
```

## El salario

Calcular el salario de una persona con base en el número de horas trabajadas y el valor de la hora.Si la persona trabaja más de 40 horas, las horas extras deben pagarse con un **incremento del 50% sobre el valor de la hora normal**.   
Al trabajador se le debe hacer una serie de **descuentos** según el rango:    
Salario hasta 200.000 un 10%   
Salario por encima de 200.000 y hasta 1.000.000 un 15%   
Salario por encima de  $1.000.000 un 18%

Mostrar el salario final y el descuento realizado.


```
print("Digite el numero de horas trabajadas")
horasTrabajadas = float(input())
print("Digite el valor de la hora")
valorHoras = float(input())
if ( horasTrabajadas > 40 ) :
  horasExtras = horasTrabajadas-40
  horassinExtras = horasTrabajadas - horasExtras 
  pagoHoraExtra = horasExtras*(valorHoras+valorHoras*0.5)
  salario = (horassinExtras*valorHoras)+pagoHoraExtra
  print("Su salario es de", salario) 
else:
  salario = horasTrabajadas*valorHoras
  print("su salario es de", salario) 
 
if ( salario <= 200000):
  descuento = salario*0.1
else:
  if (salario > 200000 and salario <= 1000000):
    descuento = salario*0.15
  else:
    if (salario > 1000000):
      descuento = salario*0.18

print("su salario final es", salario-descuento,"y su descuento es", descuento)
```
