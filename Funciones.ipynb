## Calculadora

Implementar una calculadora que sea capaz de sacar un exponente basandose solo en una función que reciba un número y devuelva el número mas 1.

```
def sumar1(x):
  y = x + 1
  return y

def sumar(x,u):
  for i in range(1,u+1):
    x = sumar1(x)
  return x

def mult(x,u):
  return x*u


a = int(input())
b = int(input())
c = sumar(a,b)
print(c)
```

## Contador de estrellas


```
import random
def imprimir(m,n):
    for i in range(0,n):
        for j in range(0,n):
            print(m[i][j], " ", end = "")
        print("")
    print("----------------------")
    
def contarEstrellas(m,n):
    c = 0
    for i in range(1,n-1):
        for j in range (1,n-1):
            if m[i][j] == 1 and m[i+1][j] == 0 and m[i-1][j] == 0 and m[i][j+1] == 0 and m[i][j-1] == 0:
                c = c + 1
    return c
                

def generarM(n):
  m = [[0 for x in range(100)] for y in range(100)]
  for i in range(0,n):
    for j in range (0,n):
        m[i][j] = random.randint(0, 1)
  return m

m = [[0 for x in range(100)] for y in range(100)]
n = 5
e = 0
contador = 0
while (e < 4):
  m = generarM(n)
  e = contarEstrellas(m,n)
  contador = contador + 1
print('Intentos',contador)
imprimir(m,n) 
print(e)
```
